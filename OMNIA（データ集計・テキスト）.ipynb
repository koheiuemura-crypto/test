#### 5-3. 集合（set）

set.jpeg

### ６. if文とインデント（字下げ）

- `if`文の基本的な構造
```
if 条件1:
    行いたい動作1  # 少し右にずらす（インデント）
elif 条件2:
    行いたい動作2  # 少し右にずらす（インデント）
elif 条件3:
    行いたい動作3  # 少し右にずらす（インデント）
・
・
else:
    上記の条件に当てはまらない時に行いたい動作  # 少し右にずらす（インデント）
```

- インデント
  - インデントとは、プログラミングの分野では、プログラムの構造を見やすくするために制御構文の内側にある行などの先頭に、一律に同じ幅の空白を挿入すること。
  - ほとんどのプログラミング言語では、インデントはコードの見た目の問題であり、プログラムの意味には影響を及ぼさない（取り除いても同じように動作する）が、Pythonはインデントによってプログラムの構造を記述する必要がある。
  - Pythonの公式では、インデントはスペース4つ分が推奨されているが、コラボはデフォルトで2つ分になっているので、2つ分でインデントしましょう。

````python
# 天気からお知らせしてくれるプログラムを実装

weather = "晴れ"  # ここを変えて実行してみましょう

if weather == "晴れ":
  print("日焼け止めを塗ってから出かけましょう。")
elif weather == "曇り":
  print("過ごしやすい天気です。")
elif weather == "雨":
  print("傘が必要です。")
else:
  print("異常気象です。")
````

#### 7-2. while文

- リスト以外（特定の条件が成り立つ間）をループさせたい時に使う。

- 気をつけないと無限ループにおちいる。無限ループに入ってしまったら、すぐに`control + c`を押すか、再生ボタン（実行ボタン）をもう一度押して、実行を強制終了して下さい！！

````python
# 1から10までをprintするプログラム

number = 1

print("プログラムスタート")

while number <= 10:
  print(number)
  number = number + 1

print("プログラム終了")
````

````python
# 無限ループ（実行する時は注意して下さい！）

number = 1

print("プログラムスタート")

while number <= 10:
  print(number)
  # number = number + 1

print("プログラム終了")
````

````python
# 例1: SyntaxError（構文を書きミスした時に出るエラー）

classmates = ["田中", "高橋", "佐藤", "佐藤", "鈴木"]

count = 0
for name in classmates:
  if nama == "佐藤":
    count = count + 1

print("このクラスには佐藤さんが" + str(count) + "人います。")
````

  - [pandas.DataFrameを結合するmerge, join（列・インデックス基準） - note.nkmk.me](https://note.nkmk.me/python-pandas-merge-join/)

````python
df_join_orders = pd.read_table(
  "/content/drive/MyDrive/OMNIA/考える/アナリティクス/データ集計・DataStorage/order_data/order_data_2022_12_04.csv",
  usecols=["order_no","driver_id","order_state","receive_type","local_area_flag"],
  delimiter=','
)

df_join_orders.head()
````

### ２. コード規則（PEP8）

- [ドキュメント](https://pep8-ja.readthedocs.io/ja/latest/)

- 構文 vs コード規則
  - 構文とは、プログラミング言語の文法や書式のことで、それらに沿った記述を行わないとエラーが発生する。
  - コード規則（コーディング規則）とは、間違いではない（エラーは出ない）が、プログラミング言語単位や会社単位、チーム単位などで決めた、コーディングする際の書き方のルール。

- なぜ間違いでないのに、コーディングに規則を設けるのか
  1. コードは書かれるより読まれる回数の方が圧倒的に多いです。他の人が読みやすく、編集/追加しやすいコード（可読性があり、汎用的なコード）を書くことを心がけましょう。
  2. 一貫性のあるコードを書いておけば、自分のケアレスミスを減らすことにも繋がります。
